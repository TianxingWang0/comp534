```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
serial = 105.815

x = np.linspace(1, 32, 32)



y = np.array([105.150, 56.162, 43.324, 34.645, 34.946, 29.850, 31.528, 28.239, 29.246, 26.935, 29.082, 27.008, 29.315, 27.377, 29.623, 27.910, 31.256, 29.939, 30.451, 28.944, 30.232, 31.412, 29.989, 28.839, 37.471, 36.014, 35.171, 33.887, 32.788, 31.169, 30.935, 30.122])

pTp = y * x

e = serial / pTp
```

```python
e
```

```python
fig = plt.figure()  

plt.ylim(0.0, 1.2)
plt.xlim(1.0, 32.0)

master, = plt.plot(x, e, '-k', linewidth = 1)  
plt.scatter(x, e, s=60, c='r', marker = '+')
plt.hlines(0.2, 0, 32, colors = "y", linestyles = "dashed")
 
plt.title('Parallel Efficiency Curve')  
plt.ylabel('Parallel Efficiency')  
plt.xlabel('Processors') 
  
# plt.legend(handles=[master],labels=['branch master'],loc='best')
 
plt.show()
```

```python
fig = plt.figure()  
ax = fig.add_subplot(111)

plt.xlim(1, 33)

ax.set_ylim(0, 1.1)
e_curve, = ax.plot(x, e, '-k', linewidth = 1, color='r')  
ax.scatter(x, e, s=60, c='blue', marker = '+')

ax2 = ax.twinx()
time_curve, = ax2.plot(x, np.flip(y),'k', linewidth = 1, color='g')
ax2.scatter(x, np.flip(y), s=60, c='r', marker = '*')

plt.title("Parallel Efficiency & Time")
ax.set_ylabel('Parallel Efficiency')  
ax2.set_ylabel('Time(s)')  
ax.set_xlabel('Processors') 
ax.hlines(0.237, 0, len(e), colors = "y", linestyles = "dashed")
  
plt.legend(handles=[e_curve, time_curve],labels=['parallel efficiency', 'time'],loc='upper center')
 
plt.show()
```
